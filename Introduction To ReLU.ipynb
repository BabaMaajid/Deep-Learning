# Activation function
The activation function is responsible for transforming the summed weighted input from the node into the activation of the node or output for that input.
# ReLU(Rectified Linear Unit) activation function is a
1) Piece wise linear activation function.

2) A node or unit that implements this activation function is known to as a rectified linear activation unit.

3) Networks that use the rectifier function for the hidden layers are referred to as rectified networks.


4) Returns the value provided as input directly, or the value 0.0 if the input is 0.0.

# Python Implemenation of Relu Function()


```python
import matplotlib.pyplot as plt
# Rectified Linear Function
def ReLU(x):
    return max(0,x)
# Empty lists for input and output
input_values =[] 
output_values = []

#List of input
for x in range(-15,20):
    input_values.append(x)
# values for output
for x in input_values:
    output_values.append(ReLU(x))
#Ploting input to the ReLU outputs
plt.plot(input_values,output_values)
plt.xlabel('inputs')
plt.ylabel('outputs')
plt.show()
```
